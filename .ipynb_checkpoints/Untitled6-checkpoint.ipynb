```python
import pandas as pd
```

```python
air_quality_no2 = pd.read_csv("air_quality_no2_long.csv",
                              parse_dates=True)


air_quality_no2 = air_quality_no2[["date.utc", "location",
                                   "parameter", "value"]]


air_quality_no2.head()
```

```python
air_quality_pm25 = pd.read_csv("air_quality_pm25_long.csv",
                               parse_dates=True)


air_quality_pm25 = air_quality_pm25[["date.utc", "location",
                                     "parameter", "value"]]


air_quality_pm25.head()
```

```python
air_quality = pd.concat([air_quality_pm25, air_quality_no2], axis=0)

air_quality.head()
```

```python
air_quality = air_quality.sort_values("date.utc")

air_quality.head()
```

```python
air_quality_ = pd.concat([air_quality_pm25, air_quality_no2], keys=["PM25", "NO2"])

air_quality_.head()
```

```python
import openaq
import pandas as pd

# Create an OpenAQ client
api = openaq.OpenAQ()

# Get locations/stations data
locations = api.locations()

# Convert to DataFrame
stations_coord = pd.DataFrame(locations['results'])

# Save to CSV file
stations_coord.to_csv('air_quality_stations.csv', index=False)

print("Air quality stations data downloaded and saved!")
stations_coord.head()
```
